**PROJECT:-1**:- Housing_price_prediction

Step:-1 . Gathering the data

```python
import numpy as np
import pandas as pd
```

```python
h_data=pd.read_csv('/content/Housing (1).csv')
```

```python
h_data.head()
```

```python
h_data.info()
```

step:2 data cleaning/data processing/data wrangling

```python
h_data['mainroad']=h_data['mainroad'].apply({'yes':0,'no':1}.get) #convert the catagorical variable into neumerical variable
```

```python
h_data.head()
```

```python
h_data['guestroom']=h_data['guestroom'].apply({'yes':0,'no':1}.get)
h_data['basement']=h_data['basement'].apply({'yes':0,'no':1}.get)
h_data['hotwaterheating']=h_data['hotwaterheating'].apply({'yes':0,'no':1}.get)
h_data['airconditioning']=h_data['airconditioning'].apply({'yes':0,'no':1}.get)
h_data['prefarea']=h_data['prefarea'].apply({'yes':0,'no':1}.get)
h_data['furnishingstatus']=h_data['furnishingstatus'].apply({'furnished':0,'semi-furnished':1,'unfurnished':3}.get)
```

```python
h_data.head()
```

**step:-3** divide the data into dependent(y) and independent(x) variable

```python
x=h_data[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']]  #these all are independent variable(i/p)
y=h_data['price']
```

```python
x.head()
```

step:-4  splitting the data into training and testing set.

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

step:5  creating ML model by using linear regression algorithm

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()  #linear regression algorithm get activated
```

```python
regression.fit(x_train,y_train)  #fit is use to train the ML learning model by providing x_train & y_train
```

```python
predictions=regression.predict(x_test)
```

```python
predictions
```

```python
regression.score(x,y)
```

```python
#convert the data from categorical to numerical
hp["mainroad"]=hp["mainroad"].apply({"yes":1,"no":0}.get)
hp["guestroom"]=hp["guestroom"].apply({"yes":1,"no":0}.get)
hp["basement"]=hp["basement"].apply({"yes":1,"no":0}.get)
hp["hotwaterheating"]=hp["hotwaterheating"].apply({"yes":1,"no":0}.get)
hp["airconditioning"]=hp["airconditioning"].apply({"yes":1,"no":0}.get)
hp["prefarea"]=hp["prefarea"].apply({"yes":1,"no":0}.get)
hp["furnishingstatus"]=hp["furnishingstatus"].apply({"furnished":1,"semi-furnished":2,"unfurnished":3}.get)
```

```python
hp.head()
```

```python
hp.info()
```

```python
hp.isnull().sum()  #for finding null values
```

```python
#devide the data into dependant and independant variable
x=hp[["area","bedrooms","bathrooms","stories","mainroad","guestroom","basement","hotwaterheating","airconditioning","parking","prefarea","furnishingstatus"]]
y=hp[["price"]]
```

```python
x.head()
y.head()
```

```python
#splitting the data into traning and testing pahse
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
x_train.shape
y_test.shape
```

```python
#creating ML model by using linear regression method
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
predictions=regression.predict(x_test)
```

```python
predictions
```

```python
regression.score(x,y)
```
