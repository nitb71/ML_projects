```python
#split data dependent and independent variable
x=data.drop("insuranceclaim",axis=1)
```

```python
y=data["insuranceclaim"]
```

```python
y.head()
```

```python
#train test split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

KNN algoritham


```python
#importing classifier algorithams
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=13)
```

```python
knn.fit(x_train,y_train)
```

```python
prediction=knn.predict(x_test)
```

```python
prediction
```

```python
#finding confusion matrices and accuracy score
from sklearn.metrics import confusion_matrix,accuracy_score
cm_knn=confusion_matrix(y_test,prediction)
ac_knn=accuracy_score(y_test,prediction)
```

```python
print(cm_knn)
```

```python
print(f"accuracy score:{ac_knn : .3f}")  #...3f use for decimal number
```

```python

from sklearn.metrics import accuracy_score

ac_scores=[ac_knn,ac_log,ac_tree,ac_forest]

models=["KNN","Logistic","Decision tree","Random forest"]
colors=["skyblue","salmon","limegreen","orange"]

plt.figure(figsize=(8,5))
bars=plt.bar(models,ac_scores,color=colors)

#for showing exact accuracy_score value above the bars
for bar,score in zip(bars,ac_scores):
  plt.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.005,f"{score:4f}",ha="center",va="bottom",fontsize=10)

plt.ylabel("accuracy_score")
plt.title("accuracy_score comparison")
plt.ylim(0.6,1.01)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
